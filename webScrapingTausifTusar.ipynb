```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.ccny.cuny.edu/registrar/fall"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
print(soup.title)
```

```python
table = soup.find("table")
rows = table.find_all("tr")
```

```python
events = []

for row in rows[1:]:
    
    cols = row.find_all("td")
    
    if len(cols) >= 3:
        
        date = cols[0].text.strip()
        dow = cols[1].text.strip()
        text = cols[2].text.strip()
        
        events.append([date, dow, text])

print(events[:5])
```

```python
df = pd.DataFrame(events, columns=["date", "dow", "text"])
df.head()
```

```python
df["date"] = [d.split("-")[0].strip() for d in df["date"]]
df["date"] = df["date"] + " 2021"
# convert to datetime
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df.head()
```

```python
df = df.set_index("date")
df.head()
```
